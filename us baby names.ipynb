MOST POPULAR NAMES IN 2018

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv("us_baby_names.csv")
```

```python
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.Gender = df.Gender.astype("category")
```

```python
names_2018 = df[df.Year == 2018].copy()
```

```python
names_2018[names_2018.Gender == "F"].nlargest(10 , "Count")
```

```python
def most_pop (year , gender, n ):
    return df[(df.Year == year) & (df.Gender == gender )].nlargest(n,"Count")
```

```python
most_pop(2018, "F", 10)
```

```python
most_pop(2018, "M", 10)
```

ADVNACED DATA AGGREGATION

```python
df.groupby(["Name","Gender"]). Count.sum().reset_index(level = -1)
```

```python
agg = df.groupby(["Name"  , "Gender"  ]).agg(Total  = ("Count", "sum"  ),No_Year = ("Count", "count" ), First_Year = ( "Year" ,"min" ) ,
                                              Last_Year =("Year","max"), Max_Count = ("Count","max"))
```

```python
agg
```

```python
def best_year(group):
    return group.nlargest(1,"Count").Year
```

```python
best_y = df.groupby(["Name","Gender"]).apply(best_year, include_groups  = False)
```

```python
best_y 
```

```python
agg.groupby("Gender").apply(lambda x: x.nlargest(10, "Total"))
```

```python
agg["Spike_Score"]    = agg.Max_Count.div(agg.Total).mul(100)
```

```python
agg
```

```python
agg.sort_values(by  = "Spike_Score", ascending = True).head(20)
```

```python
pers = agg.sort_values(by = "Spike_Score", ascending  = True). head(20)
```
